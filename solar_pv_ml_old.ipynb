```python
data = pd.concat([data2021, data2022], ignore_index=True)
# data = pd.concat([data2022], ignore_index=True)
test_data = pd.concat([data2023], ignore_index=True)
print(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.callbacks import EarlyStopping

# Select relevant columns
selected_columns = [
    '24 Hour sum solar power from solar panel (MW)'
    # '24 Hour mean solar power from solar panel (MW)'

]
train_data_selected = data[selected_columns]
test_data_selected = data2023[selected_columns]

# Split the data into training and testing sets
train_data = train_data_selected
test_data = test_data_selected

# Normalize the data
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data)
test_scaled = scaler.transform(test_data)

# Prepare the data for LSTM
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 6 # 6: Short Term, 12: Medium Term, 144: Daily cycle
X_train, y_train = create_dataset(train_scaled, train_scaled[:, 0], time_steps)
X_test, y_test = create_dataset(test_scaled, test_scaled[:, 0], time_steps)

# Define the LSTM model architecture
model = Sequential([
    LSTM(units=64, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(units=1)
])
model.compile(optimizer='adam', loss='mean_squared_error')

# Define the EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the LSTM model
history = model.fit(X_train, y_train, epochs=200, batch_size=16, validation_split=0.1, verbose=1, callbacks=[early_stopping])

# Evaluate the model performance
model.evaluate(X_test, y_test)

model.save("lstm_model.h5")

# Make predictions
predictions = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

# Plotting training and validation loss per epoch
plt.plot(history.history['loss'], label='Training Loss', linestyle='-')
plt.plot(history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss per Epoch')
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf

# Check and set GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print("GPUs available:", gpus)
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
else:
    print("No GPUs found. Running on CPU.")

# Assuming `model` is your trained LSTM model and `X_test`, `y_test` are test sets
time_horizons = [1, 3, 144, 26064]  # Number of 10-min intervals for each horizon

# Lists to store evaluation results and true/predicted values
results_list = []
true_vs_pred_data = []

# Function to calculate error metrics
def calculate_metrics(y_true, y_pred):
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
    mae = np.mean(np.abs(y_true - y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100 if np.all(y_true != 0) else np.nan
    r_squared = 1 - (np.sum((y_true - y_pred) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2))
    return rmse, mae, mape, r_squared

# Evaluate for each time horizon
for horizon in time_horizons:
    y_preds = np.zeros(len(X_test) - horizon)
    y_true_horizon = np.zeros(len(X_test) - horizon)
    
    # Loop through the test data
    for i in range(len(X_test) - horizon):
        # Predict recursively for longer horizons
        input_seq = X_test[i].reshape(1, X_test.shape[1], X_test.shape[2])
        
        for step in range(horizon):
            y_pred = model.predict(input_seq, verbose=0)  # Avoid console output
            input_seq = np.concatenate((input_seq[:, 1:, :], y_pred.reshape(1, 1, -1)), axis=1)
        
        y_preds[i] = y_pred[0, 0]
        y_true_horizon[i] = y_test[i + horizon]
        
        # Assuming you have a time index or can calculate it
        true_vs_pred_data.append({
            'Time': i * 10,  # Assuming each step represents 10 minutes
            'True': y_test[i + horizon],
            'Predicted': y_pred[0, 0],
            'Horizon': f"{horizon*10}-min"
        })

    # Calculate metrics for this horizon
    rmse, mae, mape, r_squared = calculate_metrics(y_true_horizon, y_preds)

    results_list.append({
        'Horizon': f"{horizon*10}-min",
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape,
        'R²': r_squared
    })

results_df = pd.DataFrame(results_list)
true_vs_pred_df = pd.DataFrame(true_vs_pred_data)

results_df.to_csv('evaluation_metrics.csv', index=False)
true_vs_pred_df.to_csv('true_vs_pred.csv', index=False)

print(results_df)
```
